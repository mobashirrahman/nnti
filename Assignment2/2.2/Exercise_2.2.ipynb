```python
#Implement this!
class TokenDataset(Dataset):
    def __init__(self, data_path):
        super().__init__()

        with open(data_path) as f:
            data_1 = json.load(f)  

        self.x = []
        self.y = []
        self.n_samples = len(data_1)

        for data in data_1:
            tokens = data['tokens']
            label = data['label']

            token_tensor = torch.tensor(tokens, dtype=torch.long)
            self.x.append(token_tensor)
            self.y.append(label)

        self.x = pad_sequence(self.x, batch_first=True)
        self.y = torch.tensor(self.y, dtype=torch.long)     

    def __len__(self):
        return self.n_samples
    
    def __getitem__(self, index):
        return self.x[index], self.y[index]
    


class TokenDataLoader(DataLoader):
    def __init__(self, dataset, batch_size=32, shuffle=True):
        super().__init__(dataset, batch_size=batch_size, shuffle=shuffle)
```

```python
dataset = TokenDataset("data/Dataset1/data.json")

train_size = int(0.8 * len(dataset))
test_size = len(dataset) - train_size
train_data, test_data = random_split(dataset, [train_size, test_size])

#Separate the dataset in train and test datasets at random

train_loader = TokenDataLoader(train_data, batch_size=16, shuffle=True) #Dataloader for the training part of the dataset only

for batch in train_loader:
    tokens, labels = batch  

    print(tokens[:5])
    print(labels[:5])
    
    break 
```

```python
#Implement this!
class CSVDataset(Dataset):
    def __init__(self, data_path):
        super().__init__()
        data = []
        prognosis_labels = set()
        self.encoded_data = []
        self.x = []
        self.y = []

        with open(data_path, 'r') as file:
            reader = csv.reader(file)
            headers = next(reader)  
            for row in reader:
                data.append(row)
                prognosis_labels.add(row[-1])  

        prognosis_labels = sorted(list(prognosis_labels))

        one_hot_matrix = self.one_hot_encoding(prognosis_labels)

        for row in data:
            symptoms = row[:-1]
            prognosis = row[-1]
            one_hot = one_hot_matrix[prognosis]
            encoded_row = symptoms + one_hot
            self.encoded_data.append(encoded_row)

        for row in self.encoded_data:
            features = [float(value) for value in row[:64]]
            label = [int(value) for value in row[65:]]       
            self.x.append(features)
            self.y.append(label)

        self.x = torch.tensor(self.x, dtype=torch.float32)
        self.y = torch.tensor(self.y, dtype=torch.float32) 

    def __getitem__(self, index):
        return self.x[index], self.y[index]

    def one_hot_encoding(self, prognosis_labels):
        one_hot_matrix = {}
        for i in range(len(prognosis_labels)):
            label = prognosis_labels[i]
            one_hot_vector = []
            for j in range(len(prognosis_labels)):
                if i == j:
                    one_hot_vector.append(1)
                else:
                    one_hot_vector.append(0)
            one_hot_matrix[label] = one_hot_vector

        return one_hot_matrix    

    def __len__(self):    
        return len(self.encoded_data)  


class CSVDataLoader(DataLoader):
    def __init__(self, dataset, batch_size=6, shuffle=False):
        super().__init__(dataset, batch_size=batch_size, shuffle=shuffle)
```

```python
dataset = CSVDataset("data/Dataset2/data.csv")

train_loader = CSVDataLoader(dataset, batch_size=6, shuffle=False)

for batch in train_loader:
    features, labels = batch
    print(features.shape)
    print(labels.shape)
    break  

```
